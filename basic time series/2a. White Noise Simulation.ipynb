Load Python packages

```python
import numpy as np, pandas as pd, matplotlib.pyplot as plt, statsmodels.api as sm
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
```

## 1. Simulation

Set [seed](https://www.statisticshowto.com/random-seed-definition/) for reproducibility of results

```python
np.random.seed(123)
```

Simulate white noise (100 draws from N(0,1))

```python
w_noise = np.random.normal(loc=0, scale=1, size=100)
```

Plot the draws

```python
plt.figure(figsize=(8, 3))
plt.plot(w_noise, 'o-', color='darkgreen', markersize=3, linewidth=0.6)
plt.title('White Noise', loc='left')
plt.grid(which='major', color='lightgrey', linewidth=0.3)
plt.box(False)
plt.show()
```

## 2. Autocorrelation Function (ACF) and Partial Autocorrelation Function (PACF)

```python
lag_max = 10

# initialize plot
fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(8, 3))
axs = axs.flatten()

# AC subplots
plot_acf(w_noise, lags=lag_max, alpha=0.05, zero=False, mfc='black', 
         bartlett_confint=False, title=None, ax=axs[0])
axs[0].set_ylabel('ACF')

plot_pacf(w_noise, lags=lag_max, alpha=0.05, zero=False, mfc='black', 
          title=None, ax=axs[1])
axs[1].set_ylabel('PACF')

# For all subplots set y-limits, adjust x-ticks, add x-labels
for ax in axs:
    ax.set_ylim(-0.4,0.4)
    ax.set_xticks(np.arange(1, lag_max+1, step=1))
    ax.set_xlabel('Lag')

fig.tight_layout() # Adjust layout
plt.show() # Show the plot
```
